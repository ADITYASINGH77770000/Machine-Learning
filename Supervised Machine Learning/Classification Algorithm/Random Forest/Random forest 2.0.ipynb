```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv("petrols.csv")
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
models=LabelEncoder()
transmission=LabelEncoder()
Petrol=LabelEncoder()
```

```python
df['models']=models.fit_transform(df['models'])
df['transmission']=transmission.fit_transform(df['transmission'])
df['Petrol']=Petrol.fit_transform(df['Petrol'])
```

```python
df
```

```python
input=["models","year","transmission"]
x=df[input]
y=df.Petrol
```

```python
x
```

```python
y
```

```python
from sklearn .model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df[['models','year','transmission']],df[['Petrol']],test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
reg = RandomForestClassifier(n_estimators=100,random_state=0)
reg.fit(x,y)
```

```python
reg.score(x,y)
```

```python
reg.predict(x_train)
```
