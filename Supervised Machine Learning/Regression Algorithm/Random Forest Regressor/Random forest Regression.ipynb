```python
# plotting the linear plot category and paid

import seaborn as sns 
import matplotlib.pyplot as plt 
sns.lineplot(x="Category",y="Paid",color="green",data=df)
plt.grid("darkgrid")
plt.title("linear plot")
plt.show()
```

```python
input=["Page total likes","Type","Category","Post Month","Post Weekday","Post Hour"]
x=df[input]
y=df.Paid
```

```python
x
```

```python
y.head()
```

```python
from sklearn .model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape,y_train.shape,x_test.shape,x_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor = RandomForestRegressor(n_estimators=100, random_state=0)
regressor.fit(x_train, y_train)
```

```python
regressor.score(x_train,y_train)
```

```python
regressor.score(x_test,y_test)
```

```python
regressor.predict([[139441,1,3,12,3,3]])
```

```python
from sklearn.metrics import r2_score
```

```python
x_train_prediction = regressor.predict(x_train)
r2 = r2_score(x_train_prediction , y_train)
```

```python
print(f"R^2: {r2}")
```
