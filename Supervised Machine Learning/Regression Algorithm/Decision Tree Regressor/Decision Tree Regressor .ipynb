```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv("laptop data.csv")
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
Title=LabelEncoder()
Brand=LabelEncoder()
Cpu_Type=LabelEncoder()
df['Title']=Title.fit_transform(df['Title'])
df['Brand']=Brand.fit_transform(df['Brand'])
df['Cpu_Type']=Cpu_Type.fit_transform(df['Cpu_Type'])
```

```python
df.columns
```

```python
df.drop(["Price"],axis=1,inplace=True)
```

```python
df.head()
```

```python
# plotting the linear plot between price inr and Cpu_type

import seaborn as sns 
import matplotlib.pyplot as plt 
sns.lineplot(x="Cpu_Type",y="Price_INR",color="red",data=df)
plt.grid("darkgrid")
plt.title("linear plot")
plt.show()
```

```python
inputs=['Title', 'Brand', 'Screen_Size', 'Rating', 'Ram(Gb)','Disk_Size_GB', 'Price_INR']
x=df[inputs]
y=df.Cpu_Type
```

```python
x.head()
```

```python
y.tail()
```

```python
from sklearn .model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
```

```python
regressor.fit(x_train,y_train)
```

```python
regressor.predict(x_test)
```

```python
regressor.score(x_test,y_test)
```

```python
from sklearn.metrics import r2_score
```

```python
x_train_prediction = regressor.predict(x_train)
r2 = r2_score(x_train_prediction , y_train)
```

```python
print(f"R^2: {r2}")
```
