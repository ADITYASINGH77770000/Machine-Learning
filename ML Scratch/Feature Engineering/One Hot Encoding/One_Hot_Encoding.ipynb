# OneHotEncoding using Sklearn

```python
x = df.drop(("selling_price"),axis="columns")
y = df["selling_price"]
```

```python
x.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop="first",sparse_output=False,dtype=np.int32)
```

```python
x_news = ohe.fit_transform(x[['fuel','owner']])
```

```python
x_news.shape
```

```python
x_new = np.hstack((x[['brand','km_driven']].values,x_news))
```

```python
x_new.shape
```

```python
from sklearn.model_selection import train_test_split
x_new_train,x_new_test,y_train,y_test = train_test_split(x_new,y,test_size=0.2)
```

```python
x_new_train.shape,x_new_test.shape
```

# OneHotEncoding with Top Categories

```python
counts = df['brand'].value_counts()
```

```python
df['brand'].nunique()
threshold = 100
```

```python
repl = counts[counts <= threshold].index
```

```python
pd.get_dummies(df['brand'].replace(repl, 'uncommon')).sample(5)
```
